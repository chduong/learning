### Load Necessary Libraries

```python
# Documentation Location for matplotlib: https://matplotlib.org/tutorials/introductory/pyplot.html

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
# List of values x, y required:
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y, label='2x') #label paremeter is required for plt.legend()
# plt.show() # if you show plots between .plot and .title, two graphs are generated with the data plotted on 1 graph and the title plotted on the other graph. Thus, this is commented out.

# Adding title:
plt.title('First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# Adding labels to axes:
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Adding ticks to axes:
plt.xticks(np.arange(5))
# plt.xticks([0, 1, 2, 3, 4, 100]) # tick marks automatically space out keeping order. Try it with this!
plt.yticks(np.arange(10))

plt.legend()

plt.show()
```

```python
# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=300) # If dpi=None, defaults to rcParams["savefig.dpi"] = 'figure'. If 'figure', uses the figure's dpi value.

# Plotting Line 1:
# Adding linewidth, markers/styles/sizes/colors:
# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue') # can also pass through hexadecimal colors: '#ababab'

# Shorthand notation for plot options:
# format = plt.plot(axis1, axis2, 'color', 'marker', 'line')
plt.plot(x,y, 'b^--', label='2x') # b is the color, while ^-- gives you lines made of ^--, so -^- gives you -^- lines, and .-- gives you lines made of .--

## Plotting Line 2:
# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as a line:
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2') # slice does not include 6th value.

# Plot remainder of graph as a dot:
plt.plot(x2[5:], x2[5:]**2, 'r--') # slices select segments that are lines vs dots. Overlap makes the plot look a little better/continuous.

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300) # This dpi overrides the plt.figure dpi parameter.

# Show plot
plt.show()
```

### Bar Chart

```python
# Data values and labels for chart:
labels = ['A', 'B', 'C']
values = [1,4,2]

# Parameters for the figure:
plt.figure(figsize=(5,3), dpi=100)

# The bar chart:
bars = plt.bar(labels, values)

## Add patterns to bars individually:
# bars[0].set_hatch('/') # makes first bar have / pattern

# Adding patterns iteratively:
patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```

### Load Fifa Data

```python
# Data from: https://www.kaggle.com/karangadiya/fifa19
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=bins, color='#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

labels = ['Left', 'Right']
colors = ['#abcdef', '#aabbcc']

plt.pie([left, right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart #2

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
